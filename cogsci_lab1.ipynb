8. Let's deal with the missing values. 

a) For the features EthnicGroup, ParentMaritalStatus and WklyStudyHours replace NaN values with the most common value.

b) For the feature ParentEduc replace NaN value with a string *no formal education*.

c) For the feature TestPrep replace NaN value with a string *none*.

d) For the features PracticeSport and IsFirstChild remove all the examples with NaN values. 

e) For the feature NrSiblings replace the NaN value with the median value.

f) Finally, for the feature TransportMeans replace the NaN value with a string *walking*.




```python
columns_to_replace = ['EthnicGroup', 'ParentMaritalStatus', 'WklyStudyHours']
df[columns_to_replace] = df[columns_to_replace].apply(lambda x: x.fillna(x.mode()[0]))

replacement_dict = {'ParentEduc': 'no formal education',
                    'TestPrep': 'none',
                    'TransportMeans': 'walking'}

df.fillna(replacement_dict, inplace=True)

df = df.dropna(subset=['PracticeSport', 'IsFirstChild'])

median_value = df['NrSiblings'].median()
df['NrSiblings'].fillna(median_value, inplace=True)

df
```

10. Tbhe final step to preparing the dataset for classification is transforming the non-numeric variables to numeric variables. 
For features Gender and LunchType use OneHot Encoder, and for EthnicGroup, ParentEduc, TestPrep, ParentMaritalStatus, PracticeSport, IsFirstChild, TransportMeans and WklyStudyHours use Label Encoder. 

```python
one_hot_columns = ['Gender', 'LunchType']
label_columns = ['EthnicGroup', 'ParentEduc', 'TestPrep', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild', 'TransportMeans','WklyStudyHours']

missing_columns = set(one_hot_columns + label_columns) - set(df.columns)
if missing_columns:
    raise ValueError(f"Missing columns in the DataFrame: {missing_columns}")

one_hot_encoder = OneHotEncoder()
label_encoder = LabelEncoder()
one_hot_encoder = OneHotEncoder(handle_unknown='ignore')

df_encoded_one_hot = pd.DataFrame(one_hot_encoder.fit_transform(df[one_hot_columns]).toarray(), columns=one_hot_encoder.get_feature_names_out(one_hot_columns))

df_encoded_label = df[label_columns].apply(label_encoder.fit_transform)

df_encoded = pd.concat([df_encoded_one_hot, df_encoded_label, df.drop(columns=one_hot_columns + label_columns)], axis=1)

df_encoded
```
